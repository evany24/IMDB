## Exploratory Data Analysis

### Load and Concatenate year 2000 and 2001 as Dataframes

```python
import pandas as pd
```

```python
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
df_2000.head()
```

```python
len(df_2000)
```

```python
df_2001.head()
```

```python
len(df_2001)
```

```python
df= pd.concat([df_2000, df_2001])
df.head()
```

```python
len(df)
```

### How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
df_valid = df[(df['budget'] > 0) | (df['revenue'] > 0)]
df_valid
```

```python
print(f'{len(df_valid)} Movies have budget or revenue greater than 0.')
```

### How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
cert_values = df_valid['certification'].value_counts()
cert_values
```

```python
#import numpy
import numpy as np
#seaborn
import seaborn as sns
#matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
price_fmt = StrMethodFormatter("${x:,.0f}")
```

```python
# label order
label_order = df_valid['certification'].value_counts().index
plt.figure(figsize= (10, 6))
colors = sns.color_palette("tab10")
# make MatplotLib ax object using seaborn.countplot
ax = sns.countplot(data=df_valid,
                   x='certification', 
                   order = label_order,
                   palette = colors)
# set x and y labels
ax.set_ylabel('Number of Movies', fontsize=14, weight='bold')
ax.set_xlabel('Certification', fontsize=16, weight='bold')
## Set the Title
ax.set_title('Number of movies by Certification', fontsize=16, weight='bold');
```

### What is the average revenue per certification category?

```python
avg_rev= df_valid.groupby('certification')\
    ['revenue'].agg(['mean']).sort_values('mean', ascending=False).copy()
avg_rev['mean'] = round(avg_rev['mean'],2)
avg_rev
```

```python
avg_rev = df_valid.groupby('certification')['revenue'].mean().reset_index()
avg_rev = avg_rev.sort_values(by='revenue', ascending=False)

fig, ax = plt.subplots(figsize=(10,4))
sns.barplot(data=avg_rev, x='revenue', y=avg_rev['certification']) 
plt.title('Avg Movie Revenue by Certification', fontsize = 18, weight='bold')
plt.xlabel('US Dollars', fontsize = 14, weight='bold')
plt.ylabel('Certification', fontsize = 14, weight='bold');
plt.xticks(fontsize = 8, weight='bold')
plt.yticks(fontsize = 8, weight='bold')
ax.xaxis.set_major_formatter(price_fmt)
plt.tight_layout()
plt.show();
```

### What is the average budget per certification category?

```python
bud_mean = df_valid.groupby('certification')\
            ['budget'].agg(['mean']).sort_values('mean', ascending=False).copy()
bud_mean
```

```python
bud_mean = df_valid.groupby('certification')['budget'].mean().reset_index()
bud_mean = bud_mean.sort_values(by='budget', ascending=False)

fig, ax = plt.subplots(figsize=(8,4))
sns.barplot(data=bud_mean, x='budget', y=bud_mean['certification']) 
plt.title('Avg Budget by Certification', fontsize = 18, weight='bold')
plt.xlabel('US Dollars', fontsize = 14, weight='bold')
plt.ylabel('Certification', fontsize = 14, weight='bold');
plt.xticks(fontsize = 8, weight='bold')
plt.yticks(fontsize = 8, weight='bold')
ax.xaxis.set_major_formatter(price_fmt)
plt.tight_layout()
plt.show();
```
